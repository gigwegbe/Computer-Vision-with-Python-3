## Guassian filter 

```python
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
from PIL import ImageFilter

img = Image.open("images/image.jpg")
plt.imshow(img)
plt.show()
blur_img = img.filter(ImageFilter.GaussianBlur(5))
plt.imshow(blur_img)
plt.show()
```

```python
from skimage import io
from skimage import filters
img = io.imread("images/image.jpg")
out = filters.gaussian(img, sigma=5, multichannel=True)
io.imshow(out)
io.show()
```

```python
out = filters.gaussian(img, sigma=2.5, multichannel=True)
io.imshow(out)
io.show()
```

```python
out = filters.gaussian(img, sigma=0, multichannel=True)
io.imshow(out)
io.show()
```

# Image thresholding

Thresholding means to update the color value of a pixel to either white or black based on a threshold value.
If the pixel value is greater than the threshold value, then set the pixel to "White", otherwise set it to "Black".

```python
from skimage.filters import threshold_otsu, threshold_adaptive
from skimage.io import imread, imsave
from skimage.color import rgb2gray

img = imread("images/image.jpg")
img = rgb2gray(img)

thresh_value = threshold_otsu(img)
thresh_img = img > thresh_value
```

# Edge Detection

Sobel edge detector algorithm

```python
from skimage import io
from skimage import filters
from skimage import color
img = io.imread("images/image.jpg")
img = color.rgb2gray(img)
edge = filters.sobel(img)
plt.imshow(edge)
plt.show()
```

Canny edge detector aglorithm

```python
from skimage import io
from skimage import feature
from skimage import color
img = io.imread("images/image.jpg")
img = color.rgb2gray(img)
edge = feature.canny(img, 3)
plt.imshow(edge)
plt.show()
```
